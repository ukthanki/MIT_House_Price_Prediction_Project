```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("drinks_dataset.csv")
df
```

```python
asia_data = df[df["continent"] == "Asia"]
asia_data
```

```python
asia_beer = asia_data["beer_servings"]
asia_spirits = asia_data["spirit_servings"]
asia_wine = asia_data["wine_servings"]
```

```python
asia_beer.corr(asia_spirits, method="pearson")
```

```python
asia_beer.corr(asia_wine)
```

```python
asia_data.corr(method="pearson").drop("Unnamed: 0").drop("Unnamed: 0", axis=1)
```
